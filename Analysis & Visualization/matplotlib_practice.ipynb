```python
import matplotlib.pyplot as plt 
x = ['Mon','Tue','Wed','Thu','Fri','Sat','Sun']
y1 = [13,15,16,18,15,16,12]
y2 = [15,19,14,12,15,17,12]
plt.xlabel('Daily')
plt.ylabel('Temp')
plt.plot(x,y1, label = 'Day')
plt.plot(x,y2, label = 'Night')
plt.legend(loc = 'upper right')
plt.title('Diurnal Temperature')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,5,20)
y1 = x**2
y2 = x**1.5

plt.plot(x,y1, "bo-",linewidth=4,markersize=10)
plt.plot(x,y2, "gs-",linewidth=4,markersize=10)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

a = np.random.randn(1000000)
b = 2 * np.random.randn(1000000)-1
c = 4 * np.random.randn(1000000)+2

plt.hist(a, bins=1000, density=True, alpha=0.5, histtype='step', label='(mean, stddev)=(0,1)')
plt.hist(b, bins=1000, density=True, alpha=0.75, histtype='step', label='(mean, stddev)=(-1,2)')
plt.hist(c, bins=1000, density=True, alpha=1.0, histtype='step', label='(mean, stddev)=(2,4)')

plt.show()
```
