```python
import requests
import pandas as pd
import time
from tqdm import trange
```

```python
data_list = []

draw = 1

for page in trange(0,448):
    url = f'https://news.seoul.go.kr/api/27/getCorona19Status/get_status_ajax.php?{draw}={2}'

    url = f'{url}&order[0][dir]=desc&start={page*100}&length={100}&search[value]=&search[regex]=true&_=1632482105214'

    response = requests.get(url)
    json_data = response.json()

    recordsTotal = json_data['recordsTotal']
    data = pd.DataFrame(json_data['data'])
    

    data_list.append(data)

    draw+=1

draw = 1

for page in trange(0,500):
    url = f'https://news.seoul.go.kr/api/27/getCorona19Status/get_status_ajax_pre.php?{draw}={2}'

    url = f'{url}&order[0][dir]=desc&start={page*100}&length={100}&search[value]=&search[regex]=true&_=1632482105215'

    response = requests.get(url)
    json_data = response.json()

    recordsTotal = json_data['recordsTotal']
    data = pd.DataFrame(json_data['data'])
    

    data_list.append(data)

    draw+=1
```
